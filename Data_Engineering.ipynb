```python
#importing Packages
import bs4 as bs
import urllib.request
import pandas as pd
import os
path=r"C:\Users\rchukka\Desktop\Rajesh\MachineHack"
os.chdir(path)
```

```python
#reading data as list from html pages 
Mobile_Name=[]
User_Rating=[]
Mobile_Price=[]
Total_Ratings_Reviews=[]
Spec_Ram_Rom=[]
Spec_Display=[]
Spec_Camera=[]
Spec_Battery=[]
Spec_Processor=[]
Mobile_Warranty=[]
Aval_Offers=[]
EMI_details=[]
Exchange_Price=[]
try:
    for j in range(200):
            source = urllib.request.urlopen('https://www.flipkart.com/mobiles/pr?sid=tyy,4io&otracker=categorytree&page='+str(j)).read()
            soup = bs.BeautifulSoup(source,'lxml')
            container=soup.body.find_all('div',{'class':'_1UoZlX'})
            length_container=len(container)
            #print(length_container)
            for i in range(length_container):
                    #container_title=container[0].find_all('a',{'class':'_2cLu-l'})[0]['title']
                    Mobile_Name.append(container[i].find_all('div',{'class':'_3wU53n'})[0].text)
                    #print(Mobile_Name)
                    User_Rating.append(container[i].find_all('div',{'class':'hGSR34 _2beYZw'})[0].text)
                    #print(User_Rating)
                    Mobile_Price.append(container[i].find_all('div',{'class':'_1vC4OE _2rQ-NK'})[0].text)
                    #print(Mobile_Price)
                    Total_Ratings_Reviews.append(container[i].find_all('span',{'class':'_38sUEc'})[0].text)
                    #print(Total_Ratings_Reviews)
                    Spec_Ram_Rom.append(container[i].find_all('li',{'class':'tVe95H'})[0].text)
                    #print(Spec_Ram_Rom)
                    Spec_Display.append(container[i].find_all('li',{'class':'tVe95H'})[1].text)
                    #print(Spec_Display)
                    Spec_Camera.append(container[i].find_all('li',{'class':'tVe95H'})[2].text)
                    #print(Spec_Camera)
                    Spec_Battery.append(container[i].find_all('li',{'class':'tVe95H'})[3].text)
                    #print(Spec_Battery)
                    #Spec_Processor.append(container[i].find_all('li',{'class':'tVe95H'})[4].text)
                    #print(Spec_Processor)
                    #Mobile_Warranty.append(container[i].find_all('li',{'class':'tVe95H'})[5].text)
                    #print(Mobile_Warranty)
                    #Aval_Offers.append(container[i].find_all('li',{'class':'_1ZRRx1'})[0].text)
                    #print(Aval_Offers)
                    #EMI_details.append(container[i].find_all('li',{'class':'_3MCpsc'}))
                    #print(EMI_details)
                    #Exchange_Price.append(container[i].find_all('div',{'class':'_3gVem9'})[0])
                    #print(Exchange_Price.text)
except (IndexError,ValueError):
        pass
```

```python
try:
    Flipkart_Mobile_Data=pd.DataFrame()
    Flipkart_Mobile_Data['Mobile_Name']=pd.Series(Mobile_Name).values
    Flipkart_Mobile_Data['User_Rating']=pd.Series(User_Rating).values
    Flipkart_Mobile_Data['Mobile_Price']=pd.Series(Mobile_Price).values
    Flipkart_Mobile_Data['Total_Ratings_Reviews']=pd.Series(Total_Ratings_Reviews).values
    Flipkart_Mobile_Data['Spec_Ram_Rom']=pd.Series(Spec_Ram_Rom).values
    Flipkart_Mobile_Data['Spec_Display']=pd.Series(Spec_Display).values
    Flipkart_Mobile_Data['Spec_Camera']=pd.Series(Spec_Camera).values
    Flipkart_Mobile_Data['Spec_Battery']=pd.Series(Spec_Battery).values
    #Flipkart_Mobile_Data['Spec_Processor']=pd.Series(Spec_Processor).values
    #Flipkart_Mobile_Data['Mobile_Warranty']=pd.Series(Mobile_Warranty).values
    #Flipkart_Mobile_Data['Aval_Offers']=pd.Series(Aval_Offers).values
    #Flipkart_Mobile_Data['EMI_details']=pd.Series(EMI_details).values
    #Flipkart_Mobile_Data['Exchange_Price']=pd.Series(Exchange_Price).values
except (IndexError,ValueError):
    pass
    
```
